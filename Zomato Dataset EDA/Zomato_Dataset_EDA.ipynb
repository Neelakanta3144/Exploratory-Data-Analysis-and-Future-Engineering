```python
df_country=pd.read_excel('Country-Code.xlsx')
df_country.head()
```

```python
final_df=pd.merge(df,df_country,how='left',on='Country Code')
```

```python
final_df.head()
```

```python
final_df.columns
```

```python
country_names=final_df['Country'].value_counts().index
```

```python
country_values=final_df['Country'].value_counts().values
```

```python
# Top 3 countries that uses zomato
plt.pie(country_values[:3],labels=country_names[:3],autopct='%1.2f%%')
plt.show()
```

Observation: Zomato maximum records or transactions are from India and after that there are United States and United Kingdom 

```python
ratings=final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
```

```python
ratings
```

Observations:
1. When rating b/w 4.5 to 4.9 --> Excellent
2. When rating b/w 4.0 to 4.4 --> Very good
3. When rating b/w 3.5 to 3.9 --> good
4. When rating b/w 2.5 to 3.4 --> average
5. When rating b/w 1.8 to 2.4 --> poor
6. Many people have not rated

```python
plt.figure(figsize=(12,6))
sns.barplot(x=ratings['Aggregate rating'],y=ratings['Rating Count'])
```

```python
plt.figure(figsize=(12,6))
sns.barplot(x=ratings['Aggregate rating'],y=ratings['Rating Count'],hue=ratings['Rating color'],palette=['white','red','orange','yellow','green','green'])
```

Observations:
1. Not rated has very large count
2. Maximum ratings are between 2.5 to 3.4

```python
# Count plot
sns.countplot(x='Rating color',data=ratings,palette=['blue','red','orange','yellow','green','green'])
```
